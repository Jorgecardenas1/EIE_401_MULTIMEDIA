```python
import numpy as np
from scipy.fft import fft, ifft
import matplotlib.pyplot as plt
import scipy.io.wavfile
from scipy.signal import resample

import IPython.display as ipd
from IPython.display import Audio
```

```python
#carga de audio
[sample_rate, pcm_data] = scipy.io.wavfile.read('Exercise2.7_adultmale_b_d2.wav')
amplitudes = np.array(pcm_data)

#normalización
amplitudes = 0.9*amplitudes/max(abs(amplitudes));

#plot
plt.figure(figsize=(15,3))
plt.plot(amplitudes)
plt.xlabel('Sample')
plt.ylabel('Normalized Sample')
plt.grid()

#audio
Audio(amplitudes, rate=44000)
```

```python
import scipy.signal as sps

# Your new sampling rate
new_rate = 15000

# Resample data
number_of_samples = round(len(pcm_data) * float(new_rate) / sample_rate)
print(number_of_samples)
data = sps.resample(pcm_data, number_of_samples)

#plot
plt.figure(figsize=(15,3))
plt.plot(data)
plt.xlabel('Sample')
plt.ylabel('Normalized Sample')
plt.grid()
Audio(data, rate=15000)
```
